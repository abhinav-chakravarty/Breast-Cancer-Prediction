```python
df['diagnosis'] = df['diagnosis'].map({"M":1,"B":0})
```

```python
df
```

```python
df['diagnosis'].unique()
```

```python
# diagnosis is the prediction of the disease, so it is the result of the process
# the input data will be everything apart from the diagnosis
X = df.drop('diagnosis',axis=1)
X
```

```python
Y = df['diagnosis']
Y
```

```python
# splitting the data into Train and Test, we know X = input, Y = output
# we keep 70% of the data for training and the rest for testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

### Building Machine Learning Models

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,Y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
lr_acc = accuracy_score(Y_test,y_pred)
print(lr_acc)
```

```python
# saving the accuracy for comparison with other models
results = pd.DataFrame()
tempResult = pd.DataFrame({"Algorithm":['Logistic Regression'],'Accuracy':[lr_acc]})
results = pd.concat([results,tempResult])
results = results[['Algorithm','Accuracy']]
results
```

#### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
```

```python
y_pred = dtc.predict(X_test)
y_pred
```

```python
dtc_acc = accuracy_score(Y_test,y_pred)
dtc_acc
```

```python
tempResult = pd.DataFrame({"Algorithm":['Decision Tree Classifier'],'Accuracy':[dtc_acc]})
results = pd.concat([results,tempResult])
results = results[['Algorithm','Accuracy']]
results
```

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
y_pred = rfc.predict(X_test)
y_pred
```

```python
rfc_acc = accuracy_score(Y_test,y_pred)
rfc_acc
```

```python
tempResult = pd.DataFrame({"Algorithm":['Random Forest Classifier'],'Accuracy':[rfc_acc]})
results = pd.concat([results,tempResult])
results = results[['Algorithm','Accuracy']]
results
```

#### Support Vector Classifier

```python
from sklearn import svm
svc = svm.SVC()
svc.fit(X_train,Y_train)
```

```python
y_pred = svc.predict(X_test)
y_pred
```

```python
svc_acc = accuracy_score(Y_test,y_pred)
svc_acc
```

```python
tempResult = pd.DataFrame({"Algorithm":['Support Vector Classifier'],'Accuracy':[svc_acc]})
results = pd.concat([results,tempResult])
results = results[['Algorithm','Accuracy']]
results
```
